```python
def n_of_each(iterable, n):
    for thing in iterable:
        for i in range(n):
            yield thing
```

```python
for thing in n_of_each(["a", 2, []], 3):
    print(thing)
```

```python
def n_times(iterable, n):
    for i in range(n):
        for thing in iterable:
            yield thing
```

```python
for thing in n_times(["a", 2, []], 3):
    print(thing)
```
